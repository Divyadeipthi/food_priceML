RANDOM FOREST REGRESSOR

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor  # Importing RandomForestRegressor
from sklearn.metrics import mean_squared_error
import seaborn as sns

# Define features (X) and target variable (y)
X = df_integers.drop(['mp_price'], axis=1)  # Drop the target column to get features
y = df_integers['mp_price']  # Target variable: the mp_price

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature Scaling (important for many algorithms, though Random Forests are less sensitive to scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)  # You can adjust n_estimators (number of trees)

# Train the Random Forest regressor on the training data
rf.fit(X_train_scaled, y_train)

# Make predictions on both the training and test sets
train_pred = rf.predict(X_train_scaled)
test_pred = rf.predict(X_test_scaled)

# Evaluate model performance using Mean Squared Error (MSE)
train_error = mean_squared_error(y_train, train_pred)
test_error = mean_squared_error(y_test, test_pred)

print("Train Error (MSE):", train_error)
print("Test Error (MSE):", test_error)
```

 **KNeighborsRegressor**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import seaborn as sns


# Define features (X) and target variable (y)
X = df_integers.drop(['mp_price'], axis=1)  # Drop the target column to get features
y = df_integers['mp_price']  # Target variable: the mp_price



# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature Scaling (important for KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the KNN regressor with a chosen value of K (e.g., K=5)
knn = KNeighborsRegressor(n_neighbors=5)

# Train the KNN regressor on the training data
knn.fit(X_train_scaled, y_train)

# Make predictions on both the training and test sets
train_pred = knn.predict(X_train_scaled)
test_pred = knn.predict(X_test_scaled)

# Evaluate model performance using Mean Squared Error (MSE)
train_error = mean_squared_error(y_train, train_pred)
test_error = mean_squared_error(y_test, test_pred)

print("Train Error (MSE):", train_error)
print("Test Error (MSE):", test_error)
```
